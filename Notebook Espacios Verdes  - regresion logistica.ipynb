```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
df = pd.read_csv('C:/Users/greco/Downloads/Modulo 2 - Python/Trabajo integrador EANT/completa_sin_SantaRita3.csv', sep = ';')
df_sr = pd.read_csv('C:/Users/greco/Downloads/Modulo 2 - Python/Trabajo integrador EANT/villaSantaRita3.csv', sep = ';')
```

```python
df = df.drop(['sm'], axis=1)
df_sr = df_sr.drop(['sm'], axis=1)
```

```python
df = df.drop(['TOT_POB'], axis=1)
df_sr = df_sr.drop(['TOT_POB'], axis=1)
```

```python
df = df.drop(['BARRIOS'], axis=1)
df_sr = df_sr.drop(['BARRIOS'], axis=1)
```

```python
df = df.drop(['LAYER'], axis=1)
df_sr = df_sr.drop(['LAYER'], axis=1)
```

```python
X = df.drop(['OBJETIVO'], axis=1)
y = df['OBJETIVO']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, stratify=y)
logr=LogisticRegression()
logr.fit(X, y)
```

```python
y_pred = logr.predict(X_test)
val_real = pd.Series(y_test.values)
val_pred = pd.Series(y_pred)
predicciones = pd.concat([val_real.rename('Valor real'),val_pred.rename('Valor Pred') ,abs(val_real-val_pred).rename('Dif(+/-)')] ,  axis=1)
predicciones
```
